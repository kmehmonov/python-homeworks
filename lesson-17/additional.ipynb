```python
# 1

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

num_vawes = 10 
x = np.linspace(0, 2 * np.pi, 100)
frequencies = np.linspace(1, 10, num_vawes)
time_steps = np.linspace(0, 2 * np.pi, 200)
fig, ax = plt.subplots()
lines = [ax.plot(x, np.sin(f * x))[0] for f in frequencies]


# Set axis limits and labels
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.5, 1.5)
ax.set_title("Dynamic Sinusoidal Waves")
ax.set_xlabel("x")
ax.set_ylabel("Amplitude")

# Animation update function
def update(frame):
    for line, f in zip(lines, frequencies):
        line.set_ydata(np.sin(f * x + frame))  # Update wave based on time
    return lines

# Create animation
ani = FuncAnimation(fig, update, frames=time_steps, interval=100, blit=True)

# Display the animation
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'ro')

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
plt.show()
```
