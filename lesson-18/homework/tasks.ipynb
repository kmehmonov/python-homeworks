```python
# movie.csv
# Load the CSV file into a DataFrame and display a random sample of 10 rows.


df_movie = pd.read_csv("../data/movie.csv")
df_movie.sample(10)
```

```python
# Part 2: Exploring DataFrames
# 2.1 Using the DataFrame from iris.json:
# Rename the columns to lowercase.
# Select only the sepal_length and sepal_width columns.

df_iris.columns = [item.lower() for item in df_iris.columns]
df_iris.columns
```

```python
# 2.2 From the titanic.xlsx DataFrame:
# Filter rows where the age of passengers is above 30.
# Count the number of male and female passengers (value_counts).

filtered_data = df_titanic[df_titanic.Age > 30]
sex_count = df_titanic.Sex.value_counts()
nmales = sex_count.loc["male"]
nfemales = sex_count.loc["female"]
print("number of males = ", nmales)
print("number of females = ", nfemales)

```

```python
# 2.3 From the Flights parquet file:

# Extract and print only the origin, dest, and carrier columns.
# Find the number of unique destinations.

print(df_flight[["Origin", "Dest", "CarrierDelay"]])
unique_destinations = set(df_flight["Dest"])
print(unique_destinations)
```

```python
# 2.4 From the movie.csv file:

# Filter rows where duration is greater than 120 minutes.
# Sort the filtered DataFrame by director_facebook_likes in descending order.

filtered_rows = df_movie[df_movie["duration"] > 120]
filtered_rows.sort_values(by="director_facebook_likes", ascending=False)
```

```python
# From iris.json: Calculate the mean, median, and standard deviation
#  for each numerical column.

numeric_iris = df_iris.select_dtypes('number')

print("means")
means = numeric_iris.mean(axis=0)
print(means)

print("medians")
medians = numeric_iris.median(axis=0)
print(medians)

print("stdv")
stdv = numeric_iris.std(axis=0)
print(stdv)

```

```python
# From titanic.xlsx: Find the minimum, maximum, and sum of passenger ages

print("max age of passanger is", df_titanic.Age.max(axis=0))
print("max age of passanger is", df_titanic.Age.min(axis=0))
print("max age of passanger is", df_titanic.Age.sum(axis=0))
```

```python
# From movie.csv:

# Identify the director with the highest total director_facebook_likes.
# Find the 5 longest movies and their respective directors.

most_liked_director = df_movie.sort_values(by="director_facebook_likes").loc[0, "director_name"]
longest_movies = df_movie.sort_values(by="duration").iloc[0:5, 1]
print("most_liked_director: ", most_liked_director)
print("longest_movies:", longest_movies)
```

```python
# From Flights parquet file:
# Check for missing values in the dataset.
#  Fill missing values in a numerical column with the column’s mean.

for column in df_flight.select_dtypes('number').columns:
    df_flight[column].fillna(df_flight[column].mean(), inplace=True)

```
