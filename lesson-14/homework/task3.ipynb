```python
import json
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

# Open Demoblaze homepage
driver.get("https://www.demoblaze.com")
time.sleep(2)  # Allow time for the page to load

# Click on "Laptops" section
laptops_section = driver.find_element(By.LINK_TEXT, "Laptops")
laptops_section.click()
time.sleep(2)  # Allow time for the page to load

# Function to extract laptop data
def extract_laptops():
    laptops = []
    items = driver.find_elements(By.CSS_SELECTOR, ".card-block")
    for item in items:
        try:
            name = item.find_element(By.CSS_SELECTOR, "h4.card-title > a").text
            price = item.find_element(By.CSS_SELECTOR, "h5").text
            description = item.find_element(By.CSS_SELECTOR, "p.card-text").text
            laptops.append({
                "name": name,
                "price": price,
                "description": description
            })
        except Exception as e:
            print(f"Error extracting item: {e}")
    return laptops

# List to store all laptops data
all_laptops = []

# Loop through pages
while True:
    all_laptops.extend(extract_laptops())
    try:
        next_button = driver.find_element(By.ID, "next2")
        if not next_button.is_enabled():
            break
        next_button.click()
        time.sleep(2)  # Allow time for the next page to load
    except Exception as e:
        print("No more pages or error:", e)
        break

# Close the browser
driver.quit()

# Save data to JSON file
with open("laptops.json", "w") as file:
    json.dump(all_laptops, file, indent=4)

print(f"Scraped {len(all_laptops)} laptops and saved to laptops.json.")
```
