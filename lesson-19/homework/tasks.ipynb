```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
# Merging and Joining
# task-1

with sqlite3.connect(r'data\chinook.db') as con:
    df_customers = pd.read_sql("select * from customers", con=con)
    df_invoices = pd.read_sql("select * from invoices", con=con)

merged = pd.merge(df_customers, df_invoices, on='CustomerId', how="inner")
result = (
    merged.groupby(["CustomerId", "FirstName", "LastName"])
    .size()
    .reset_index(name="TotalInvoices")
    .sort_values(by="TotalInvoices", ascending=False)
)
result


```

```python
# Grouping and Aggregating
# task-1

df = pd.read_excel(r'data/titanic.xlsx')

result = (
    df.groupby('Pclass')
    .agg(
        Age=('Age', 'mean'),
        Fare=('Fare', 'sum'),
        passenger_count=('PassengerId', 'count')
    )
)

print(result)
```

```python
# task-2

df = pd.read_csv(r'data/movie.csv')
grouped = df.groupby(['color', 'director_name'])

result = (
    grouped.agg(
        num_critic_for_reviews=('num_critic_for_reviews', 'count' ),
        average_duration=('duration', 'mean')
        )
)
result

```

```python
# task-3

df = pd.read_parquet(r'data/flights')
```

```python
# task-3 continue

df['ArrDelay'] = pd.to_numeric(df['ArrDelay'], errors='coerce')
df['DepDelay'] = pd.to_numeric(df['DepDelay'], errors='coerce')
df['Month'] = pd.to_numeric(df['Month'], errors='coerce')
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

result = (
    df.groupby(['Year', 'Month'])
    .agg(
        total_flights=('Flights', 'count'),
        avg_arrival_delay=('ArrDelay', 'mean'),
        max_departure_delay=('DepDelay', 'max') 
    )
    .sort_values(by=["Year", "Month"])
)
result

```

```python
# Applying Functions
# task-1

def classification(item):
    if item < 18:
        return "Child"
    else:
        return "Adult"

df = pd.read_excel(r'data/titanic.xlsx')

df["Age_Group"] = df["Age"].apply(classification)
df.head(2)
```

```python
# task- 2

def normalize(group):
    group['Salary'] = (group.Salary-group.Salary.min())/(group.Salary.max()-group.Salary.min())
    return group

with sqlite3.connect(r'data/chinook.db') as con:
    df_employee = pd.read_sql("select * from employees", con=con)

df_normalized = df_employee.groupby("ReportsTo").apply(normalize)
df_normalized.head()
```

```python
# task-3

def classifier(series):
    if series < 60:
        return "short"
    elif series < 120:
        return "medium"
    else:
        return "long"


df_movie = pd.read_csv('data/movie.csv')

df_movie["custom_category"] = df_movie["duration"].apply(classifier)

df_movie
```

```python
# Using pipe
# task-1

df = pd.read_excel('data/titanic.xlsx')


result = (df
            .pipe(lambda df: df[df['Survived'] == 1])
            .pipe(lambda df: df.assign(Age = df['Age'].fillna(df['Age'].mean())))
            .pipe(lambda df: df.assign(Fare_Per_Age=df['Fare'] / df['Age']))
        )

result.head()
```

```python
# task-2 

df.columns.tolist()
df['DepDelayMinutes'] = pd.to_numeric(df['DepDelayMinutes'], errors='coerce')
df['DepTime'] = pd.to_numeric(df['DepTime'], errors='coerce')
df['ArrTime'] = pd.to_numeric(df['ArrTime'], errors='coerce')
result = (
    df
    .pipe(lambda df: df[df['DepDelayMinutes'] > 30])
    .pipe(lambda df: df.assign(Delay_Per_Hour = df['DepDelayMinutes'] / (df['ArrTime']-df['DepTime'])))

)
result
```
